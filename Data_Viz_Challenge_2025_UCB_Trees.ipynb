```python
import pandas as pd
```

```python
ds=pd.read_csv('/content/Data_Viz_Challenge_2025-UCB_Trees.csv')
```

```python
ds
```

```python
ds.info()
```

```python
# Fill missing species names with 'Unknown'
ds['Species'].fillna('Unknown', inplace=True)
```

```python
# Fill missing common names with 'Unknown'
ds['Common Name'].fillna('Unknown', inplace=True)
```

```python
# Fill missing Canopy Spread and Height with median values (to maintain realistic values)
ds['Canopy Spread'].fillna(ds['Canopy Spread'].median(), inplace=True)

```

```python
ds['Height'].fillna(ds['Height'].median(), inplace=True)
```

```python
# Ensure data types are correct
ds['Tree ID'] = ds['Tree ID'].astype(str)  # Treat Tree ID as a string to avoid unnecessary calculations
```

```python
ds
```

```python
# Save cleaned dataset for Tableau
cleaned_file_path = "Data_Viz_Challenge_2025-UCB_Trees_Cleaned.csv"
ds.to_csv(cleaned_file_path, index=False)

print(f"Cleaned dataset saved as: {cleaned_file_path}")
```
